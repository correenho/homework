```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_data_complete.head()
```

```python
school_num = school_data_complete['School ID'].nunique()
student_num = school_data_complete['Student ID'].count()
undup_budget_df = school_data_complete.loc[:,['School ID','budget']].drop_duplicates()
budget_sum = undup_budget_df['budget'].sum()
avg_math = school_data_complete['math_score'].mean()
avg_read = school_data_complete['reading_score'].mean()
ptg_pass_math = school_data_complete.math_score[school_data_complete.math_score >= 70].count() / school_data_complete.math_score.count()*100
ptg_pass_read = school_data_complete.reading_score[school_data_complete.reading_score >= 70].count() / school_data_complete.reading_score.count()*100
ptg_passing = (avg_math+avg_read)/2
dist_sum = pd.DataFrame({'Total Schools':school_num,'Total Students':student_num,'Total Budget':budget_sum,'Average Math Score':avg_math,'Average Reading Score':avg_read,'% Passing Math':ptg_pass_math,'% Passing Reading':ptg_pass_read,'% Overall Passing Rate':ptg_passing},index=[0])
dist_sum
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#School Summary
school_group_df = school_data_complete.groupby('school_name')

sch_name = school_data_complete.school_name.unique()
sch_type = list(school_data.sort_values(by='school_name').type)
ttl_student = school_data_complete['size'].unique()
ttl_budget = school_data_complete.budget.unique()
per_stu_bud = [i/j for i,j in zip(ttl_budget,ttl_student)]
sch_avg_math = school_group_df.math_score.mean()
sch_avg_read = school_group_df.reading_score.mean()

ttl_math_pass = school_data_complete[school_data_complete['math_score'] >= 70]
group_math_pass = ttl_math_pass.groupby('school_name')
sch_math_pass = list(group_math_pass.school_name.count()/school_group_df['size'].count()*100)

ttl_read_pass = school_data_complete[school_data_complete['reading_score'] >= 70]
group_read_pass = ttl_read_pass.groupby('school_name')
sch_read_pass = list(group_read_pass.school_name.count()/school_group_df['size'].count()*100)

sch_ovl_pass_rate = [(i+j)/2 for i,j in zip(sch_math_pass, sch_read_pass)]

sch_sum_df = pd.DataFrame({'School Type':sch_type,
                           'Total Students':ttl_student, 
                           'Total Budget':ttl_budget,
                           'Per Student Budget':per_stu_bud, 
                           'Avg Math Score': sch_avg_math, 
                           'Avg Reading Score': sch_avg_read,
                           '% Passing Math':sch_math_pass, 
                           '% Passing Reading':sch_read_pass,
                           'Overall Passing Rate':sch_ovl_pass_rate})
sch_sum_df
```

## Top Performing Schools (By Passing Rate)

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Math Scores by Grade
school_data_renamed = school_data_complete.rename(columns={'school_name':'School Name'})
sch_9th = school_data_renamed.loc[school_data_renamed.grade=='9th', ['School Name','math_score']]
sch_10th = school_data_renamed.loc[school_data_renamed.grade=='10th', ['School Name','math_score']]
sch_11th = school_data_renamed.loc[school_data_renamed.grade=='11th', ['School Name','math_score']]
sch_12th = school_data_renamed.loc[school_data_renamed.grade=='12th', ['School Name','math_score']]

sch_9th_math_mean = sch_9th.groupby(by='School Name').math_score.mean()
sch_10th_math_mean = sch_10th.groupby(by='School Name').math_score.mean()
sch_11th_math_mean = sch_11th.groupby(by='School Name').math_score.mean()
sch_12th_math_mean = sch_12th.groupby(by='School Name').math_score.mean()

sch_grade_math_summary = pd.DataFrame({'9th':sch_9th_math_mean,'10th':sch_10th_math_mean,'11th':sch_11th_math_mean,
                                       '12th':sch_12th_math_mean})

pd.options.display.float_format = '{:,.2f}'.format
sch_grade_math_summary
```
