```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mse_data = pd.read_csv(mouse_drug_data_to_load)
clc_trial_data = pd.read_csv(clinical_trial_data_to_load)
mse_data.head(),clc_trial_data.head()
# Combine the data into a single dataset
combined_data = pd.merge(mse_data,clc_trial_data,on='Mouse ID',how = 'inner')

# Display the data table for preview
combined_data.head(10)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame
dropped = combined_data.drop(columns = ['Metastatic Sites'])
avg_tumor_vol = dropped.groupby(by=['Drug','Timepoint'],as_index=False).mean()
avg_tumor_vol = pd.DataFrame(avg_tumor_vol)
avg_tumor_vol.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
# Preview DataFrame

std_err = combined_data.groupby(['Drug','Timepoint']).sem()['Tumor Volume (mm3)']
std_err = pd.DataFrame(std_err).reset_index()
std_err.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked
avg_tumor_vol_reformatted = avg_tumor_vol.pivot(index = 'Timepoint',columns = 'Drug')['Tumor Volume (mm3)']
avg_tumor_vol_reformatted.head()
```

```python
# Generate the Plot (with Error Bars)
# Save the Figure
std_err_reformatted = std_err.pivot(index = 'Timepoint',columns = 'Drug')['Tumor Volume (mm3)']
plt.errorbar(avg_tumor_vol_reformatted.index, avg_tumor_vol_reformatted['Capomulin'], yerr=std_err_reformatted['Capomulin'],
            color = 'r', marker = 'o', linestyle = 'dashed')
plt.errorbar(avg_tumor_vol_reformatted.index, avg_tumor_vol_reformatted['Infubinol'], yerr = std_err_reformatted['Infubinol'],
            color = 'b', marker = '^', linestyle = 'dashed')
plt.errorbar(avg_tumor_vol_reformatted.index, avg_tumor_vol_reformatted['Ketapril'], yerr = std_err_reformatted['Ketapril'],
            color = 'g', marker = 's', linestyle = 'dashed')
plt.errorbar(avg_tumor_vol_reformatted.index, avg_tumor_vol_reformatted['Placebo'], yerr = std_err_reformatted['Placebo'],
            color = 'black', marker = 'D', linestyle = 'dashed')
plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.legend(['Capomulin','Infubinol','Ketapril','Placebo'],fontsize='small')
plt.grid(axis='y')

plt.savefig('Output/tumor_response_chart_w_err_bars.png')
```

```python
# Show the Figure
plt.show()
```

![Tumor Response to Treatment](../Images/treatment.png)

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metastatic_grouped = combined_data.groupby(by=['Drug','Timepoint'], as_index=False)['Metastatic Sites'].mean()
# Convert to DataFrame
metastatic_df = pd.DataFrame(metastatic_grouped)
# Preview DataFrame
metastatic_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
std_err_m = combined_data.groupby(['Drug','Timepoint']).sem()['Metastatic Sites']
# Convert to DataFrame
std_err_m_df = pd.DataFrame(std_err_m).reset_index()
# Preview DataFrame
std_err_m_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastatic_df_reformatted = metastatic_df.pivot(index = 'Timepoint',columns = 'Drug')['Metastatic Sites']
# Preview that Reformatting worked
metastatic_df_reformatted.head()
```

```python
# Generate the Plot (with Error Bars)
# Save the Figure
# Show the Figure
std_err_m_pivot = std_err_m_df.pivot(index = 'Timepoint', columns = 'Drug')['Metastatic Sites']

plt.errorbar(metastatic_df_reformatted.index, metastatic_df_reformatted['Capomulin'], yerr=std_err_m_pivot['Capomulin'],
            color = 'r', marker = 'o', linestyle = 'dashed')
plt.errorbar(metastatic_df_reformatted.index, metastatic_df_reformatted['Infubinol'], yerr = std_err_m_pivot['Infubinol'],
            color = 'b', marker = '^', linestyle = 'dashed')
plt.errorbar(metastatic_df_reformatted.index, metastatic_df_reformatted['Ketapril'], yerr = std_err_m_pivot['Ketapril'],
            color = 'g', marker = 's', linestyle = 'dashed')
plt.errorbar(metastatic_df_reformatted.index, metastatic_df_reformatted['Placebo'], yerr = std_err_m_pivot['Placebo'],
            color = 'black', marker = 'D', linestyle = 'dashed')

plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treatment Duration (Days)')
plt.ylabel('Met. Sites')
plt.legend(['Capomulin','Infubinol','Ketapril','Placebo'],fontsize='small')
plt.grid(axis='y')

plt.savefig('Output/met_chart_w_err_bars.png')

plt.show()
```

![Metastatic Spread During Treatment](../Images/spread.png)

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count=combined_data.groupby(['Drug','Timepoint'])['Mouse ID'].count()
# Convert to DataFrame
mouse_count_df = pd.DataFrame(mouse_count).reset_index()
# Preview DataFrame
mouse_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count_pivot = mouse_count_df.pivot(index='Timepoint',columns='Drug')['Mouse ID']
# Preview the Data Frame
mouse_count_pivot.head(12)
```

```python
# Generate the Plot (Accounting for percentages)

plt.plot(mouse_count_pivot['Capomulin']/.25, color='r',marker = 'o', linestyle = 'dashed', markersize='4',linewidth='1')
plt.plot(mouse_count_pivot['Infubinol']/.25, color='b',marker = '^', linestyle = 'dashed',markersize='4',linewidth='1')
plt.plot(mouse_count_pivot['Ketapril']/.25, color='g',marker = 's', linestyle = 'dashed',markersize='4',linewidth='1')
plt.plot(mouse_count_pivot['Placebo']/.25, color='black',marker = 'D', linestyle = 'dashed',markersize='4',linewidth='1')

plt.title('Survival During Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rate%')
plt.grid(axis='both')
plt.legend(['Capomulin','Infubinol','Ketapril','Placebo'],fontsize='small',loc=3)
# Save the Figure

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
# Display the data to confirm
ptg_chg_calc = (avg_tumor_vol_reformatted.iloc[9,:]/avg_tumor_vol_reformatted.iloc[0,:]-1)*100
ptg_chg_calc
```

```python
# Store all Relevant Percent Changes into a Tuple
# Splice the data between passing and failing drugs
# Orient widths. Add labels, tick marks, etc. 
ptg_chg = ptg_chg_calc.sort_values()
data_4_chart = ptg_chg[['Capomulin','Infubinol','Ketapril','Placebo']]

plt.bar(data_4_chart.index, data_4_chart, width = -1, align='edge',
        color=(data_4_chart > 0).map({True:'r',False:'g'}))

plt.grid(axis='both')
plt.title('Tumor Change Over 45 Days of Treatment')
plt.ylabel('% Tumor Volume Change')

# Use functions to label the percentages of changes
x_axis=[]
tick_values=[]

def label(x,value):
    x_axis = [x for x in range(len(data_4_chart))]
    tick_values = ['{:.0%}'.format(round(value,0)/100) for value in data_4_chart]
    for x, value in zip(x_axis, tick_values):
        plt.text(x, 0, value,horizontalalignment='right')

# Call functions to implement the function calls
label(x_axis, tick_values)

# Save the Figure
plt.savefig('Output/summary_bar_chart.png')

# Show the Figure
plt.show()
```
